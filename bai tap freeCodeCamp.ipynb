```python
s = ["32 + 698", "3801 - 2", "45 + 43", "123 + 49"]
def arithmetic_arranger(problems, solve = False):
    
    s = [problem.split() for problem in problems]

    def arranged_problems(s) :
        line1 = ''
        line2 = ''
        line3 = ''
        line4 = ''
     

        if len(s) > 5:
            return "Error: Too many problems."
        else:        
           
            for lst in s:
                if lst[1] not in ["+","-"]:
                    return "Error: Operator must be '+' or '-'."
                elif lst[0].isnumeric() == False or lst[2].isnumeric()==False:
                    return "Error: Numbers must only contain digits."
                elif len(lst[0]) > 4 or len(lst[2]) > 4 :
                    return "Error: Numbers must only contain digits."
                else:
                    kq = eval("".join(lst))     
                    ez = str(abs(kq))              
                    n = max(len(ez), max([len(c) for c in lst])) + 2
                    if n > 6: n = 6
                    line1 += abs((len(lst[0]) - n)) * " " + lst[0] + ' ' * 4
                    line2 += lst[1] + abs((len(lst[2]) - n + 1)) * " " + lst[2] + ' ' * 4
                    line3 += n * '-' + ' ' * 4
                    if kq > 0:
                        line4 += abs((len(ez) - n)) * ' ' + ez + ' ' * 4
                    else:
                        line4 += '-' + (abs((len(ez) - n)) - 1) * ' ' + ez + ' ' * 4
                            
        res = line1[:-4] + '\n' + line2[:-4]  + '\n' + line3[:-4] 
        if solve: res += '\n' + line4[:-4] 
        
        return res
    
    return arranged_problems(s)

arithmetic_arranger(["32 - 698", "1 - 3801", "45 + 43", "123 + 49"], True)
    
    


    
```

```python
class Category:
    
    def __init__(self, name):
        self.name = name
        self.ledger = []
        self.amount, self.totalDep,self.totalWith,  = 0, 0, 0, 
        
    def deposit(self, amount, description=None):
        self.amount += amount
        self.description = description if description else ''
        self.ledger.append({'amount': amount, 'description': self.description})
        self.totalDep += amount

    def withdraw(self, amount, description=None):
        self.amount -= amount       
        self.description = description if description else ''
        if self.amount < 0:
            self.amount += amount
            return False
        else:
            self.totalWith += amount
            self.ledger.append({'amount': - amount, 'description': self.description})
            return True

    def get_balance(self):

        return self.totalDep - self.totalWith

    def transfer(self, amount, des):
        if self.amount >= amount:
            self.amount -= amount
            des.amount += amount
            self.ledger.append({'amount': self.amount, 'description': 'Transfer to' + des.name})
            des.ledger.append({'amount': des.amount, 'description': 'Transfer from' + self.name})
            return True
        else:
            return False

    def check_funds(self, amount):
        return self.amount < amount

    def __str__(self):
        self.title = self.name.center(30, '*')
        for c in self.ledger:
            self.title += '\n' + f"{c['description']:<23} {c['amount']:>6}"
            return self.title

food = Category('food')
f = food.deposit(213123123, 'adfsdf')

str(food)




```
